```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
data="D:\\NLP\\session-1\\android-games - android-games.csv"
android_df=pd.read_csv(data)
android_df
```

```python
android_df.isnull().sum()
```

```python
#wordcloud is one of the method to show most repeated words in pictorical forms
# most repeated word displayed as more zoom
# less repeated words  as zom  out
# you need to selct a column
# column has different rows
# you need to combine the rows
```

```python
text=' '.join(cat.split()[1] for cat in android_df.category)
text
```

```python
from wordcloud import WordCloud
word_cloud=WordCloud(height=4000,width=4000,background_color='white')
word_cloud=word_cloud.generate(text)
plt.imshow(word_cloud)
plt.title("most common repeated words in positive customer comments")
plt.axis("off")
plt.show()
```
