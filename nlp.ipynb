```python
def text_clean(text1):
    words=text1.split()
    print(words)
    re_punct=re.compile('[%s]'%re.escape(string.punctuation))
    w=[re_punct.sub('',w) for w in words]
    w=[i.lower() for i in w]
    return w
```

```python
#load the data
file_name="D:\\NLP\\session-1\\metamorphosis clean.txt"
file=open(file_name,'rt',encoding="UTF-8-sig")
text=file.read()
#split into tkenizers
tokens=word_tokenize(text)
#convert into lower case
tokens=[w.lower() for w in tokens]
#prepeare regex by removing the puncutataions
re_punct=re.compile('[%s]'%re.escape(string.punctuation))
#remove the puncuatations
stripped=[re_punct.sub('',w) for w in tokens]
#will remove all tokens thata re not alphanumeric
words=[word for word in stripped if word.isalpha()]
#remove the stopwords
stop_words=stopwords.words("english")
words=[w for w in words if  w not in stop_words]
print(words[:10])
from nltk.stem.porter import PorterStemmer
porter=PorterStemmer()
stem_words=[porter.stem(word) for word in words]
print(stem_words[:10])
```
